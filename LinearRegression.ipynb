```python
import pandas as pd
import numpy as np
```

```python
#Taking in data
column_label = ['X','Y']
data = pd.read_csv('ex1data1.txt',names = column_label)
```

```python
data
```

```python
x = data['X']
y = data['Y']
x
```

```python
x0 = np.ones(len(x))
x0
```

```python
new = pd.DataFrame([x0,x])
```

```python
new = new.T
m = len(new)
new
```

```python
#Initializing theta and assigning a learning rate (alpha)
theta = np.zeros(len(new.columns))
theta = pd.Series(theta)
alpha = 0.01
theta
```

```python
#The cost function is a representation of the error
def cost_func(X,y,theta):
    return (1/len(X))*sum((X.dot(theta) - y))
```

```python
#Gradient is the derivative of the cost function and allows us to reach a global minimum
def gradient(X,y,theta):
    return X.T.dot((X.dot(theta) - y))
```

```python
iterations = int(input("How many iterations do you want the program to do?: "))
```

```python
#More iterations = Better the accuracy
for i in range(1500):
    theta = theta - alpha * (1/m) * gradient(new,y,theta)
```

```python
#TESTING using the initial theta.
temp = np.zeros(len(new.columns))
print(cost_func(new,y,temp))
print(cost_func(new,y,theta))
```
